```python
import tensorflow as tf
from tensorflow import keras
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#LOad FMNIST dataset
```

```python
fashion_mnist = keras.datasets.fashion_mnist

(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
#check what dataset contains
```

```python
x_train.shape
```

```python
#normalize data now
```

```python
x_train = x_train / 255 
x_test = x_test / 255
```

```python
#Show one image
```

```python
plt.imshow(x_train[0], cmap='grey')
plt.show()
```

```python
#Check its label
```

```python
y_train[0]
```

```python
#create NEURAL Network
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128,
                       activation='relu'),
    keras.layers.Dense(10,
                       activation='softmax')
])
```

```python
#Compile the Model
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Train Model
```

```python
history = model.fit(x_train, y_train,
                    epochs=5)
```

```python
# evaluate model accurcy on test data

model.evaluate(x_test, y_test)
```

```python
# predict labels for test images

pred = model.predict(x_test)
```

```python
#display one test image

plt.imshow(x_test[5], cmap= 'grey')
plt.show()
```

```python
# priinnt predicted class label

print("predicted:", np.argmax(pred[5]))
print("actual:",y_test[5])
```
