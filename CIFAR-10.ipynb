```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
#doenload dataset
(x_train, y_train), (x_test, y_test) =keras.datasets.cifar10.load_data()
```

```python
# check what dataset contain
x_train.shape
```

```python
# normalize dataset

x_train = x_train / 255
x_test = x_test / 255
```

```python
# display onee image
plt.imshow(x_train[0])
plt.show()
```

```python
#Check its label
y_train[0]
```

```python
#create NEURAL Network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(32,32,3)),
    keras.layers.Dense(128,
                       activation='relu'),
    keras.layers.Dense(10,
                       activation='softmax')
])
```

```python
#Compile the Model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Train Model
history = model.fit(x_train, y_train,
                    epochs=12)
```

```python
print(x_test.shape)
print(y_test.shape)
```

```python
# evaluate model accurcy on test data

model.evaluate(x_test, y_test)
```

```python

# predict labels for test images

pred = model.predict(x_test)
```

```python
# priinnt predicted class label

print("predicted:", np.argmax(pred[5]))
print("actual:",y_test[5])
```

```python
plt.imshow(x_test[4], cmap= 'grey')
plt.show()

print("predicted:", np.argmax(pred[4]))
print("actual:",y_test[4])

```
